# Exploratory Data Analysis (EDA) Steps

## 1. Memeriksa Informasi Dasar Data
- **Tinjauan Struktur Data**: Gunakan `.info()` untuk melihat tipe data, memeriksa kolom yang mungkin perlu diperbaiki, dan mengidentifikasi data yang kosong atau tidak sesuai.
- **Distribusi Missing Values**: Identifikasi kolom yang mengandung nilai kosong untuk menentukan apakah perlu di-impute atau dihapus.
- **Statistik Dasar**: Gunakan `.describe()` untuk menganalisis statistik deskriptif dari setiap kolom numerik, seperti rata-rata, median, dan standar deviasi.

## 2. Analisis Kolom Kategorikal
- **Frekuensi Jenis Perjalanan**: Tampilkan distribusi frekuensi untuk kolom seperti `Trip_type` (misalnya, Hail vs. Dispatch) untuk memahami perilaku penggunaan.
- **Pengguna Berdasarkan Hari atau Waktu**: Analisis kolom `Pickup_day`, `Pickup_hour`, dan `Pickup_month` untuk menemukan pola waktu yang mungkin relevan bagi bisnis, seperti jam atau hari dengan permintaan tertinggi.
- **Distribusi Zona Pickup/Drop-off**: Gunakan `Pickup_location` dan `Dropoff_location` untuk mengetahui zona mana yang sering digunakan.

## 3. Distribusi Kolom Numerik
- **Durasi dan Jarak Perjalanan**: Buat histogram atau box plot untuk `Trip_duration` dan `Trip_distance` guna mengidentifikasi pola umum atau outliers.
- **Analisis Tarif**: Periksa distribusi `Fare_amount` untuk memahami pola harga rata-rata dan mengenali apakah ada tarif ekstrem yang memerlukan perhatian khusus.

## 4. Hubungan Antar Fitur (Analisis Korelasi)
- **Heatmap Korelasi**: Buat heatmap korelasi antar kolom numerik seperti `Trip_duration`, `Trip_distance`, dan `Fare_amount` untuk melihat hubungan linier yang signifikan.
- **Scatter Plot**: Visualisasikan `Trip_duration` vs. `Trip_distance` untuk melihat pola durasi terkait jarak.

## 5. Analisis Waktu (Time Series Analysis)
- **Perjalanan Berdasarkan Jam atau Hari**: Buat time series berdasarkan `Pickup_hour` atau `Pickup_day` untuk melihat tren penggunaan pada waktu tertentu (misalnya, jam sibuk vs. jam sepi).
- **Perjalanan Berdasarkan Bulan**: Tinjau pola bulanan untuk melihat tren musiman atau perubahan terkait liburan atau cuaca.

## 6. Analisis Outlier
- **Boxplot untuk Outlier dalam Tarif dan Jarak**: Gunakan boxplot pada kolom `Fare_amount` dan `Trip_distance` untuk mendeteksi outlier yang bisa mempengaruhi hasil analisis.
- **Pengecekan Outlier Berdasarkan Periode Waktu**: Periksa apakah terdapat outlier pada jam tertentu (misalnya, kenaikan tarif ekstrem pada jam sibuk).

## 7. Pengelompokan Lokasi (Geospatial Analysis)
- **Distribusi Lokasi Pickup dan Drop-off**: Buat scatter plot atau heatmap geospasial untuk visualisasi pickup/drop-off yang sering.
- **Clustering Zona Pengguna**: Gunakan teknik clustering (misalnya, K-Means) untuk mengelompokkan zona pickup/drop-off yang mirip.

## 8. Segmentasi Pengguna
- **Klasifikasi Jenis Pengguna**: Berdasarkan variabel seperti `Trip_type`, `Fare_amount`, dan `Trip_distance`, buat segmen pelanggan (misalnya, pengguna perjalanan pendek vs. jauh).
- **Perilaku Berdasarkan Metode Pembayaran**: Jika ada data `Payment_method`, analisis pembayaran tunai vs. non-tunai untuk memahami preferensi pengguna.


# Data Analysis Workflow

## 1. Data Collection (Pengumpulan Data)
- **Tujuan**: Mengumpulkan data yang akan dianalisis, baik dari sumber internal seperti basis data, API, atau file CSV, maupun sumber eksternal seperti layanan data publik.
- **Output**: Data mentah dalam bentuk tabel, dataset, atau data file.

## 2. Data Understanding
- **Tujuan**: Memahami konteks dan struktur data yang diperoleh sebelum mulai membersihkan atau menganalisisnya.
- **Langkah-langkah**:
  - Memeriksa Struktur Data: Gunakan `.info()`, `.head()`, atau `.describe()` untuk mendapatkan gambaran awal.
  - Pemeriksaan Variabel: Identifikasi tipe variabel, jumlah kolom numerik, kategorikal, dan melihat distribusi setiap kolom.
- **Output**: Pemahaman awal mengenai kualitas, ukuran, dan isi data.

## 3. Data Cleaning (Pembersihan Data)
- **Nama Lain**: Data Sanitization, Data Cleansing
- **Tujuan**: Menangani data yang tidak lengkap, duplikat, atau tidak sesuai yang dapat mengganggu analisis.
- **Langkah-langkah**:
  - Menangani Missing Values: Menggunakan imputasi, penghapusan, atau metode lainnya untuk mengisi nilai yang hilang.
  - Menghapus atau Menggabungkan Data Duplikat: Menghapus data yang berulang untuk menghindari bias.
  - Perbaikan Inkonsistensi Data: Memperbaiki kesalahan input seperti format tanggal, ejaan yang tidak konsisten, atau tipe data yang salah.
- **Output**: Dataset yang sudah bersih dan siap untuk diolah lebih lanjut.

## 4. Data Wrangling (Transformasi Data)
- **Nama Lain**: Data Munging
- **Tujuan**: Mengubah struktur atau format data agar sesuai untuk analisis.
- **Langkah-langkah**:
  - Feature Extraction: Memisahkan atau menggabungkan kolom (misalnya, ekstrak bulan dan tahun dari tanggal).
  - Data Transformation: Melakukan scaling, encoding data kategorikal, atau transformasi logaritmik jika diperlukan.
  - Handling Outliers: Mengidentifikasi dan menangani outliers agar tidak mempengaruhi hasil analisis.
- **Output**: Dataset dalam format yang siap untuk eksplorasi dan analisis statistik.

## 5. Exploratory Data Analysis (EDA)
- **Nama Lain**: Data Profiling
- **Tujuan**: Menganalisis data untuk menemukan pola, distribusi, anomali, dan wawasan awal.
- **Langkah-langkah**:
  - Visualisasi Distribusi: Menggunakan histogram, boxplot, atau violin plot untuk melihat distribusi data.
  - Analisis Korelasi: Membuat heatmap korelasi untuk memeriksa hubungan antara variabel numerik.
  - Analisis Kategorikal: Melihat frekuensi atau proporsi untuk variabel kategorikal.
- **Output**: Wawasan mengenai karakteristik data yang membantu menentukan langkah-langkah selanjutnya.

## 6. Feature Engineering (Rekayasa Fitur)
- **Nama Lain**: Feature Creation
- **Tujuan**: Membuat atau memodifikasi fitur agar model lebih mampu menangkap pola dalam data.
- **Langkah-langkah**:
  - Transformasi Fitur: Melakukan polynomial transformation, log transformation, atau binning.
  - Fitur Interaksi: Membuat fitur berdasarkan interaksi antar kolom (contoh: `x1 * x2`).
  - Feature Encoding: Menggunakan encoding seperti one-hot encoding untuk variabel kategorikal.
- **Output**: Dataset dengan fitur tambahan atau yang dimodifikasi yang bisa meningkatkan performa model.

## 7. Data Preparation (Persiapan Data)
- **Nama Lain**: Data Preprocessing
- **Tujuan**: Mempersiapkan data sebelum model training, seperti splitting dan scaling.
- **Langkah-langkah**:
  - Data Splitting: Membagi data menjadi train dan test set, atau train-validation-test jika diperlukan.
  - Scaling & Normalization: Melakukan standardization atau normalization untuk kolom numerik agar berada dalam skala yang serupa.
- **Output**: Dataset yang sudah dibagi dan dinormalisasi sehingga siap digunakan untuk pelatihan model.

## 8. Modeling (Pemodelan)
- **Nama Lain**: Model Training, Machine Learning Model Training
- **Tujuan**: Membangun model berdasarkan data training dan mengoptimalkan performanya.
- **Langkah-langkah**:
  - Pemilihan Model: Memilih jenis model berdasarkan jenis data dan tujuan analisis (misalnya, regresi, klasifikasi, clustering).
  - Training Model: Melatih model menggunakan data training.
  - Model Evaluation: Mengevaluasi model menggunakan metrik performa seperti akurasi, F1 score, atau R-squared.
- **Output**: Model yang terlatih dan dioptimalkan untuk prediksi atau analisis lebih lanjut.

## 9. Model Evaluation (Evaluasi Model)
- **Nama Lain**: Model Validation
- **Tujuan**: Mengevaluasi performa model menggunakan data test dan metrik performa.
- **Langkah-langkah**:
  - Cross-Validation: Menggunakan teknik seperti K-fold cross-validation untuk memastikan generalisasi model.
  - Pemeriksaan Overfitting/Underfitting: Memeriksa apakah model terlalu sesuai dengan data training atau kurang mampu menangkap pola data.
- **Output**: Metrik evaluasi yang menunjukkan kekuatan dan kelemahan model.

## 10. Deployment & Monitoring (Penggunaan dan Pemantauan Model)
- **Nama Lain**: Model Serving, Operationalization
- **Tujuan**: Mengimplementasikan model dalam lingkungan produksi dan memantau kinerjanya secara real-time.
- **Langkah-langkah**:
  - Deployment: Mempersiapkan model untuk digunakan di aplikasi atau sistem tertentu.
  - Monitoring: Memantau performa model di produksi untuk menangani perubahan distribusi data atau degradasi performa.
- **Output**: Model yang diterapkan dalam sistem dan dimonitor untuk performa jangka panjang.
